# Impute with statistical values

```python
# Impute missing values
# Replace missing values in the Age column with the mean value

# calculate mean
mean_age = df['Age'].mean()
print('Mean age: {}'.format(int(mean_age)))

# imputation
df['Age'].fillna(int(mean_age), inplace=True)

# check again
print('Num of Missing values in "Age" : {}'.format(df['Age'].isnull().sum()))
```

```python
# Explore the distribution of the Age column
age_range = int(df['Age'].max() - df['Age'].min())

sns.histplot(df['Age'], bins=age_range, kde=False)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Replace missing values in the Embarked column with the mode value

# calculate mode
mode_embark = df['Embarked'].mode()[0]
print(mode_embark)

# imputation
df['Embarked'].fillna(mode_embark, inplace=True)

# check again
print('Num of Missing values in "Embarked" : {}'.format(df['Embarked'].isnull().sum()))
```

#  (Extra) Using ML method to fill in missing values:
## Predicting missing values of age

X_train: All columns except data["Age"] with non null values

y_train: data["Age"] with non null values

X_test: All columns except data["Age"] features with null values

y_test (y_pred): data["Age"] with null values

```python
from sklearn.linear_model import LinearRegression

data = pd.read_csv("titanic.csv")
```

```python
# feature selection, filter out columns of passengerID, name, embarked.
data = data[["Survived", "Pclass", "Sex", "SibSp", "Parch", "Fare", "Age"]]

# convert column sex to categorical value
data["Sex"] = [1 if x=="male" else 0 for x in data["Sex"]]

# train test split
test_data = data[data["Age"].isnull()]

data.dropna(inplace=True)
y_train = data["Age"]
X_train = data.drop("Age", axis=1)
X_test = test_data.drop("Age", axis=1)

# train model
model = LinearRegression()
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)
```

```python
test_data['Age'] = y_pred
```

```python
complete_data = pd.concat([data, test_data])

age_range = int(df['Age'].max() - df['Age'].min())
plt.hist(complete_data["Age"], bins=age_range, edgecolor="k", alpha=0.7)
plt.title("Distribution of Age After Imputation")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
